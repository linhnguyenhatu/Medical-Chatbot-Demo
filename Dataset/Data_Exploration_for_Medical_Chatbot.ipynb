```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import numpy as np
```

```python
data = pd.read_csv('sample_data/train_medical_chatbot.csv', on_bad_lines='skip')
```

```python
data['word_count_patient'] = data['Patient'].apply(lambda x: len(str(x).split(" ")))
data['word_count_doctor'] = data['Doctor'].apply(lambda x: len(str(x).split(" ")))
```

```python
#Word Cloud of Patient Prompts
all_prompt = ' '.join(text for text in data['Patient'].astype(str))
wordcloud_patient = WordCloud(width = 800, height = 400, background_color = 'white').generate(all_prompt)

plt.figure(figsize = (10, 5))
plt.imshow(wordcloud_patient)
plt.axis('off')
plt.title('Word Cloud of Patient Prompts')
plt.show()
```

```python
#Word Cloud of Doctor Responses
all_responses = ' '.join(text for text in data['Doctor'].astype(str))
wordcloud_doctor = WordCloud(width = 800, height = 400, background_color = 'white').generate(all_responses)

plt.figure(figsize = (10, 5))
plt.imshow(wordcloud_doctor)
plt.axis('off')
plt.title('Word Cloud of Doctor Responses')
plt.show()
```

```python
#Word Count Patient Box Plot

sns.boxplot(x = data['word_count_patient'], color = 'Green')
plt.xlabel('Word Count')
plt.ylabel('Patient Prompts')
plt.title('Distribution of Word Counts of Patient Prompts')
plt.show()
```

```python
#Word Count Patient Box Plot

sns.boxplot(x = data['word_count_doctor'], color = 'Blue')
plt.xlabel('Word Count')
plt.ylabel('Doctor Responses')
plt.title('Distribution of Word Counts of Doctor Responses')
plt.show()
```

```python
#Violin Plot of Word Count of Patient Prompts

plt.figure(figsize = (9, 6))
sns.violinplot(y = data["word_count_patient"], color = 'Purple')
plt.xlabel('Word Count Frequency')
plt.ylabel('Patient Prompts')
plt.title('Distribution of Word Counts of Patient Prompts')
plt.show()
```

```python
#Violin Plot of Word Count of Patient Prompts

plt.figure(figsize = (9, 6))
sns.violinplot(y = data["word_count_doctor"], color = 'Orange')
plt.xlabel('Word Count Frequency')
plt.ylabel('Doctor Responses')
plt.title('Distribution of Word Counts of Doctor Responses')
plt.show()
```
